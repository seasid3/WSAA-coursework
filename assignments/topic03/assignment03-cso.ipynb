# Exchequer Account (Historical Series)

This program retrieves the dataset for the "exchequer account (historical series)" from the CSO and stores it into a file called "cso.json".

```python
import requests
import json

url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en"

# Define a function to get data from the API
def get_all():
    response = requests.get(url)
    return response.json()

# Save the data into a JSON file 
# Code from Topic 03 writing code video
# Indent code from questioning claude AI on how to make the output look nice in the JSON file
# (as Alt Shift F formats python code, not JSON content inside a file)
if __name__ == "__main__":
    with open("cso.json", "wt") as fp:
        print(json.dumps(get_all(), indent=2), file=fp)
```
