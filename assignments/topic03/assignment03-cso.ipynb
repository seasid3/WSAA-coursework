# Exchequer Account (Historical Series)

This program retrieves the dataset for the "exchequer account (historical series)" from the CSO and stores it into a file called "cso.json".

```python
import requests

url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en"

# Define a function to get data from the API
def getdata():
    response = requests.get(url)
    return response.json()

# Test the function
if __name__ == "__main__":
    data = getdata()
    print(data)
```
